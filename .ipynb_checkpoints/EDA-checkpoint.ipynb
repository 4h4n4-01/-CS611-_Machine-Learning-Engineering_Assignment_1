```python
# Imports
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

spark = SparkSession.builder \
    .appName("EDA") \
    .config("spark.sql.shuffle.partitions", "8") \
    .getOrCreate()

print("Spark started")
```

```python
# Loading the tables from silver layer
attr = spark.read.parquet("datamart/silver/silver_feature_attributes")
fin = spark.read.parquet("datamart/silver/silver_feature_financials")
cs = spark.read.parquet("datamart/silver/silver_feature_clickstream")
loan = spark.read.parquet("datamart/silver/silver_lms_loan_daily")

print("Attributes:", attr.count(), "rows,", len(attr.columns), "cols")
print("Financials:", fin.count(), "rows,", len(fin.columns), "cols")
print("Clickstream:", cs.count(), "rows,", len(cs.columns), "cols")
print("Loans:", loan.count(), "rows,", len(loan.columns), "cols")
```

```python
print("Attributes Schema")
attr.printSchema()

print("Financials Schema")
fin.printSchema()
```

```python
loan_pd = loan.toPandas()
loan_pd["installment_num"] = loan_pd["installment_num"].astype(int)
loan_pd["tenure"] = loan_pd["tenure"].astype(int)
loan_pd["overdue_amt"] = loan_pd["overdue_amt"].astype(float)

# Will give the last installment per loan
final = loan_pd[loan_pd["installment_num"] == loan_pd["tenure"]]
final = final.copy()
final["is_default"] = (final["overdue_amt"] > 0).astype(int)

print("Default Rate:", round(final["is_default"].mean() * 100, 2), "%")

# Plot
final["is_default"].value_counts().plot(kind="bar", color=["steelblue", "red"])
plt.title("Default vs Non-Default")
plt.xticks([0, 1], ["No Default", "Default"], rotation=0)
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
# Checking age distribution of Customers
attr_pd = attr.toPandas()

plt.figure(figsize=(8, 4))
sns.histplot(attr_pd["Age"].dropna(), bins=20, color="steelblue")
plt.title("Customer Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
# Checking top 10 occupations of customers
attr_pd["Occupation"].value_counts().head(10).plot(kind="barh", color="teal")
plt.title("Top 10 Occupations")
plt.xlabel("Count")
plt.tight_layout()
plt.show()
```

```python
# Confirming the financial features and how they are distributed 
fin_pd = fin.toPandas()

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

sns.histplot(fin_pd["Annual_Income"].dropna(), bins=30, ax=axes[0], color="steelblue")
axes[0].set_title("Annual Income Distribution")

sns.histplot(fin_pd["Outstanding_Debt"].dropna(), bins=30, ax=axes[1], color="coral")
axes[1].set_title("Outstanding Debt Distribution")

sns.histplot(fin_pd["Credit_Utilization_Ratio"].dropna(), bins=30, ax=axes[2], color="green")
axes[2].set_title("Credit Utilization Ratio")

plt.tight_layout()
plt.show()
```

```python
# Credit Mix Breakdown 
fin_pd["Credit_Mix"].value_counts().plot(kind="bar", color=["green", "orange", "red"])
plt.title("Credit Mix Distribution")
plt.xticks(rotation=0)
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
# Checking for missing values 
print("Missing Values in Financials")
missing = fin_pd.isnull().sum()
missing = missing[missing > 0]
print(missing)

missing.plot(kind="bar", color="coral")
plt.title("Missing Values per Column")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
spark.stop()
print("Done")
```
